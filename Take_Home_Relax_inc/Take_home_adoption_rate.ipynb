```python
#read user data
dft = pd.read_csv('takehome_users.csv', encoding = "ISO-8859-1")
```

```python
# read engagement df
dfe = pd.read_csv('takehome_user_engagement.csv', encoding = "ISO-8859-1")
```

```python
# subtract 7 days to the date (to add the next 7 days together)
dfe['Date'] = pd.to_datetime(dfe['time_stamp']) - pd.to_timedelta(7, unit='d')
```

```python
dfe.head()
```

```python
#create a week group
week_group = dfe.groupby(['user_id', pd.Grouper(key='Date', freq='W-MON')])['visited'].sum().reset_index()
```

```python
#filter by > 2 visits
week_group[week_group['visited'] > 2]
```

```python
# create binary feature
week_group['adopted_user'] = [1 if x > 2 else 0 for x in week_group['visited']]
```

```python
week_group[week_group['visited'] > 2]
```

```python
# truncate columns
adopted_user = week_group[week_group['visited'] > 2][['user_id', 'adopted_user']]
```

```python
adopted_user
```

```python
# create unique list
adopted_user = adopted_user.groupby('user_id')['adopted_user'].mean().reset_index()
```

```python
adopted_user
```

```python
dft.head()
```

```python
# merge to main dataframe
dft = dft.merge(adopted_user, left_on='object_id', right_on='user_id', how='outer')
```

```python
dft
```

```python
dft
```

```python
# fill 0s for non merged values
dft['adopted_user'] = dft['adopted_user'].fillna(0)
```

```python
dft
```

```python
# count the adopted users
adopted_grp = dft.groupby('adopted_user')['adopted_user'].count()
```

```python
adopted_grp
```

```python
# create graph to show adoption rate
plt.figure()
sns.barplot(adopted_grp.index,adopted_grp.values)
plt.title("adoption rate among users (1 = adopted user)")
plt.savefig('adoption_rate.png')
plt.show()
```

```python
# create dummy variables for creation source
dft = pd.concat([dft, pd.get_dummies(dft['creation_source'])], axis=1)
```

```python
# drop unessessary catagories
dft.drop(columns=['name','email','creation_source','user_id','SIGNUP_GOOGLE_AUTH'], inplace=True)
```

```python
X = dft
```

```python
X
```

```python
# create org_id feature
org_id_group = X.groupby('org_id')['object_id'].count().reset_index()
```

```python
# rename the feature
org_id_group.rename(columns={'object_id': "num_in_group"},inplace=True)
```

```python
org_id_group
```

```python
# merge to the main dataframe
X = X.merge(org_id_group, left_on='org_id', right_on='org_id', how='outer')
```

```python
X
```

```python
# create user_invites feature - number of users that have been invited by specific user
user_invites = X.groupby('invited_by_user_id')['object_id'].count().reset_index()
```

```python
# rename the feature
user_invites.rename(columns={'object_id': "user_invites"},inplace=True)
```

```python
user_invites
```

```python
# merge onto main dataframe
X = X.merge(user_invites, left_on='invited_by_user_id', right_on='invited_by_user_id', how='outer')
```

```python
X
```

```python
# fill NA values
X[['user_invites', 'invited_by_user_id']] = X[['user_invites', 'invited_by_user_id']].fillna(-1)
```

```python
X
```

```python
X['create_time_rank'] = X['creation_time'].rank(method='max')
```

```python
X
```

```python
# drop unnessesary columns
X.drop(columns=['creation_time', 'last_session_creation_time'], inplace=True)
```

```python
X
```

```python
# add target
y = X['adopted_user']
```

```python
# drop target
X.drop(columns=['adopted_user'], inplace=True)
```

```python
# import libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
# split out the training set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# creat lr object
lr = LogisticRegression()
```

```python
# fit the data
lr.fit(X_train, y_train)
```

```python
# predict the data
a = lr.predict(X_test)
```

```python
# determine accuracy score
accuracy_score(a, y_test)
```

```python
# print out the lr coefficients
dict(zip(list(X_train.columns), lr.coef_[0]))
```
